```python
#import libraries for ML
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#split and read the dataset
fileData = pd.ExcelFile('/content/Case 3 - Demand Management.xlsx')
voda_data_1 = pd.read_excel(fileData, 1 )
voda_data_2 = pd.read_excel(fileData, 2)
voda_data_3 = pd.read_excel(fileData,3)
voda_data_4 = pd.read_excel(fileData, 4)
voda_data_5 = pd.read_excel(fileData ,5)
voda_data_6 = pd.read_excel(fileData, 6)
voda_data_7 = pd.read_excel(fileData, 7)
voda_data_8 = pd.read_excel(fileData, 8)
```

```python
# Delete the first two rows of every DataFrame and shift cells up
voda_data_1_cleaned = voda_data_1.drop([0, 1])
voda_data_2_cleaned = voda_data_2.drop([0, 1])
voda_data_3_cleaned = voda_data_3.drop([0, 1])
voda_data_4_cleaned = voda_data_4.drop([0, 1])
voda_data_5_cleaned = voda_data_5.drop([0, 1])
voda_data_6_cleaned = voda_data_6.drop([0, 1])
voda_data_7_cleaned = voda_data_7.drop([0, 1])
voda_data_8_cleaned = voda_data_8.drop([0, 1])
```

```python
# Rename the first column as 'Employee Code' and set it as the identifier
voda_data_1_cleaned.rename(columns={'Unnamed: 0': 'Employee Code'}, inplace=True)
voda_data_2_cleaned.rename(columns={'Unnamed: 0': 'Employee Code'}, inplace=True)
voda_data_3_cleaned.rename(columns={'Unnamed: 0': 'Employee Code'}, inplace=True)
voda_data_4_cleaned.rename(columns={'Unnamed: 0': 'Employee Code'}, inplace=True)
voda_data_5_cleaned.rename(columns={'Unnamed: 0': 'Employee Code'}, inplace=True)
voda_data_6_cleaned.rename(columns={'Unnamed: 0': 'Employee Code'}, inplace=True)
voda_data_7_cleaned.rename(columns={'Unnamed: 0': 'Employee Code'}, inplace=True)
voda_data_8_cleaned.rename(columns={'Unnamed: 0': 'Employee Code'}, inplace=True)
# Reset index to easily navigate rows
voda_data_1_cleaned.reset_index(drop=True, inplace=True)
voda_data_2_cleaned.reset_index(drop=True, inplace=True)
voda_data_3_cleaned.reset_index(drop=True, inplace=True)
voda_data_4_cleaned.reset_index(drop=True, inplace=True)
voda_data_5_cleaned.reset_index(drop=True, inplace=True)
voda_data_6_cleaned.reset_index(drop=True, inplace=True)
voda_data_7_cleaned.reset_index(drop=True, inplace=True)
voda_data_8_cleaned.reset_index(drop=True, inplace=True)
```

Reshape the first sheet


```python
# Initialize lists to store the calculated metrics
employee_ids = []
total_contracted_hours = []
total_allocated_hours = []
total_remaining_capacity = []
average_utilization_rate = []
# Iterate through the DataFrame to extract and calculate the values
i = 0
while i < len(voda_data_1_cleaned):
    if "EMP_" in str(voda_data_1_cleaned.loc[i, "Employee Code"]):
        emp_id = voda_data_1_cleaned.loc[i, "Employee Code"]
        contracted_hours = voda_data_1_cleaned.iloc[i + 1, 1:].astype(float).sum()
        allocated_hours = voda_data_1_cleaned.iloc[i + 2, 1:].astype(float).sum()
        remaining_capacity = contracted_hours - allocated_hours
        utilization_rate = (allocated_hours / contracted_hours) * 100 if contracted_hours > 0 else 0

        # Append the results
        employee_ids.append(emp_id)
        total_contracted_hours.append(contracted_hours)
        total_allocated_hours.append(allocated_hours)
        total_remaining_capacity.append(remaining_capacity)
        average_utilization_rate.append(utilization_rate)

        # Skip the next rows (including tasks)
        i += 3
    else:
        i += 1
print(  employee_ids)
print(total_contracted_hours)
print(total_allocated_hours)
print(total_remaining_capacity)
print(average_utilization_rate)
```

```python
# Create the DataFrame with the correct structure
central_usa_df = pd.DataFrame({
    'Employee Code': employee_ids,
    'Total Contracted Hours': total_contracted_hours,
    'Total Allocated Hours': total_allocated_hours,
    'Total Remaining Capacity': total_remaining_capacity,
    'Average Utilization Rate': average_utilization_rate
})
central_usa_df = central_usa_df.sort_values(by='Employee Code')
central_usa_df.reset_index(drop=True, inplace=True)
print(central_usa_df.head)
print(central_usa_df.shape)
```

```python
# Save central_usa_df data to a new Excel file
central_usa_df.to_excel('central_usa_df.xlsx', index=False)
```
